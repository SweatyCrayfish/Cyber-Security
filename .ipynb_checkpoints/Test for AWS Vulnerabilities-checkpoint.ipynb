```python
# Import the necessary libraries
import boto3

# Set the AWS region and credentials
region = "us-east-1"
access_key = "your_access_key"
secret_key = "your_secret_key"

# Create a session with the AWS credentials
session = boto3.Session(
    region_name=region,
    aws_access_key_id=access_key,
    aws_secret_access_key=secret_key
)

# Create an AWS Inspector client
inspector = session.client("inspector")

# List all the assessment templates
templates = inspector.list_assessment_templates()["assessmentTemplateArns"]

# Create a function to test for vulnerabilities
def test_vulnerabilities(template_arn):
  # Start an assessment run using the specified template
  run = inspector.start_assessment_run(
      assessmentTemplateArn=template_arn,
      assessmentRunName="TestRun"
  )
  
  # Wait for the assessment run to complete
  while True:
    status = inspector.describe_assessment_runs(
        assessmentRunArns=[run["assessmentRunArn"]]
    )["assessmentRuns"][0]["state"]
    
    if status == "COMPLETED":
      break
    elif status == "ERROR":
      print("Error running assessment")
      return
      
  # Get the findings from the assessment run
  findings = inspector.list_findings(
      assessmentRunArns=[run["assessmentRunArn"]],
      filter={
          "severities": ["High", "Medium", "Low"]
      }
  )
  
  # Print the findings
  for finding in findings["findingArns"]:
    print(inspector.describe_findings(findingArns=[finding]))

# Test for vulnerabilities using each assessment template
for template in templates:
  test_vulnerabilities(template)
```
