```python
# Import the necessary libraries
import sqlite3

# Set the name of the database file
db_file = "database.db"

# Connect to the database
conn = sqlite3.connect(db_file)

# Define a function to test for XSS vulnerabilities
def test_xss(input_str):
  # Insert the input string into the database
  with conn:
    conn.execute("INSERT INTO inputs (value) VALUES (?)", (input_str,))
  
  # Query the database for the input string
  with conn:
    result = conn.execute("SELECT value FROM inputs WHERE value=?", (input_str,)).fetchone()
  
  # Check if the input string is returned unmodified
  if result and result[0] == input_str:
    print("Not vulnerable:", input_str)
  else:
    print("Vulnerable to XSS:", input_str)

# Test for XSS using common payloads
payloads = [
  "<script>alert('XSS')</script>",
  "<img src='x' onerror='alert(1)'>",
  "<a href='javascript:alert(1)'>Link</a>",
]

for payload in payloads:
  test_xss(payload)
```

```python
# Import the necessary libraries
import sqlite3
from bs4 import BeautifulSoup

# Set the name of the database file
db_file = "database.db"

# Connect to the database
conn = sqlite3.connect(db_file)

# Define a function to test for XSS vulnerabilities
def test_xss(input_str):
  # Insert the input string into the database
  with conn:
    conn.execute("INSERT INTO inputs (value) VALUES (?)", (input_str,))
  
  # Query the database for the input string
  with conn:
    result = conn.execute("SELECT value FROM inputs WHERE value=?", (input_str,)).fetchone()
  
  # Use BeautifulSoup to parse the HTML response
  soup = BeautifulSoup(result[0], "html.parser")
  
  # Check if the input string is returned unmodified
  if soup.get_text() == input_str:
    print("Not vulnerable:", input_str)
  else:
    print("Vulnerable to XSS:", input_str)

# Test for XSS using common payloads
payloads = [
  "<script>alert('XSS')</script>",
  "<img src='x' onerror='alert(1)'>",
  "<a href='javascript:alert(1)'>Link</a>",
]

for payload in payloads:
  test_xss(payload)
```
